## `for` 반복문을 써볼까요?

#### `for` 은 다음과 같이 씁니다. 한국어로는 `풔` 이라 읽고 `--한동안` 라고 뜻을 풀이 합니다. 
#### 프로그램할때 다음과 같이 씁니다. 그리고 자동으로 1씩 더해집니다. 
```julia
for *변수* in *시작* : * 끝*
    *할일*
end

while *조건*
    *할일*
end





for *변수* in *시작* : * 끝*
    for *변수* in *시작* : * 끝*
    *할일*
    end
end
```

````julia
for j in 1:5 
    for i in 1:5
        print(i*j,",")
     end
end
````

### `for` 에서는 더해주지 않아도 바로 더해지면서 계산합니다. 
#### 다음은 `while` 에서 썻던 예제를 `for` 반복문 으로 바꾸어봤습니다. 

````julia
내가좋아하는친구 = ( 정세헌 = [2010,"James",'😉'] , 장서진 = [2011,"Mimi",'😸'], 강대한 = [2011,"DHKang",'👦'])
i = 1
for i in 1:length(내가좋아하는친구) #length 는 안에 몇개가 있는지 데려옵니다.
    내친구별명 =  내가좋아하는친구[i][2]
    내친구아이콘 = 내가좋아하는친구[i][3]
    println(""" 안녕 $내친구별명'` 님 오늘 하루도 "즐거운하루 되세요~" $내친구아이콘 """)
end
````

## for문은 왜쓸까요?
### 중첩된구문을 쓸때 더 간결하게 쓸수있습니다. 
### 중첩이란 한개의 반복문 아래 또다른 반복문을 쓰는것입니다. 


````julia
i=0
j=0
for i in 1:3
    for j in 1:3
        print("j=",j,",")
    end
    println("i=",i)
end
````

## 이렇게 해도 됩니다. 

````julia
i=0
j=0
for i in 0:1, j in 0:1 
    print(j," + ",i," = " )
    println(j + i)
end
````

````julia
i=0
j=0
C = [i + j for i in 0:1, j in 0:1] # 요렇게는 어레이에서만 됩니다.
print(C)
````

## `🤺중첩어레이내공` : 이것은 구구단같은 작은 몹은 한번에 킬합니다. 
```julia
*상위변수*=0;*하위변수*=0;[*할일* for *상위변수* in *상위변수의 시작숫자*:*상위변수가 끝날숫자* , *하위변수* in *하위변수의 시작숫자*:* 하위변수가 끝날숫자*]
```

````julia
# 구구단 은 실습을 해보세요~!
i=0;j=0; C= [i * j for i in 1:9, j in 1:9]
````
